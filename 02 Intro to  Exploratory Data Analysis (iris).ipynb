# Diagramas de Caja 

Los diagramas de caja son una manera conveniente de representar gráficamente grupos de datos numéricos a través de sus cuartiles. Las parcelas de caja también pueden tener líneas que se extienden verticalmente desde las cajas (bigotes) que indican la variabilidad fuera de los cuartiles superior e inferior, de ahí los términos gráficos de caja. Los valores atípicos se pueden representar como puntos individuales. Las parcelas de caja no son paramétricas: muestran variación en las muestras de una población estadística sin hacer ninguna suposición de la distribución estadística subyacente. 

Los espaciamientos entre las diferentes partes de la caja indican el grado de dispersión (dispersión) y asimetría en los datos, y muestran valores atípicos.  Los diagramas de caja se pueden dibujar horizontalmente o verticalmente (Tukey, 1977)

# Datos Atípicos 

Se encontraron datos atípicos en la anchura de sépalo tanto abajo del límite inferior como
superior, se procederá a un análisis considerando la variable especie

En estadística, un valor atípico es un punto de observación distante de otras observaciones. Un
valor atípico puede deberse a variabilidad en la medición o puede indicar error experimental; Estos
últimos a veces son excluidos del conjunto de datos. Los valores atípicos pueden ocurrir por
casualidad en cualquier distribución, pero a menudo indican error de medición o que la población
tiene una distribución de cola pesada. En el primer caso, se desea descartarlos o utilizar estadísticas
robustas a los valores atípicos, mientras que en el último caso indican que la distribución tiene una
alta asimetría y que se debe ser muy cauteloso en el uso de herramientas o intuiciones que asuman
una distribución normal. Una causa frecuente de valores atípicos es una mezcla de dos
distribuciones, que pueden ser dos subpoblaciones distintas.

# PDF(Función de densidad de probabilidad)

La PDF es la densidad de probabilidad más que la masa de probabilidad. El concepto es muy similar a la densidad de masa en la física. 

La función de densidad de probabilidad (PDF) es una expresión estadística que define una distribución de probabilidad para una variable aleatoria continua. Cuando se representa gráficamente la PDF, el área bajo la curva indicará el intervalo en el que la variable caerá.

 El área total en este intervalo del gráfico es igual a la probabilidad de que se produzca una variable aleatoria continua.
