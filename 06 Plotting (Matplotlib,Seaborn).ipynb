En estadstica, la regresin polinmica es una forma de anlisis de regresin en la que la relacin entre la variable independiente x y la variable dependiente y se modela como un polinomio de ensima potencia en x. La regresin polinmica se ajusta a una relacin no lineal entre el valor de x y la correspondiente media condicional de y, denominada E. Aunque la regresin polinmica ajusta un modelo no lineal a los datos, como problema de estimacin estadstica es lineal, en el sentido de que la funcin de regresin E es lineal en los parmetros desconocidos que se estiman a partir de los datos. Por esta razn, la regresin polinmica se considera un caso especial de regresin lineal mltiple. Las variables explicativas resultantes de la expansin polinmica de las variables de "lnea de base" se conocen como trminos de alto grado

```python
credit = pd.read_csv('CasoCreditos.csv')
```

```python
credit.head()
```

```python
convert_dictionary= {
    'loan_id':str,
    'status':str,
    'due_date':str,
    'paid_date':str,
    'loan_amount':int,
    'risk_score':float,
    'gender':str,
    'age':float,
    'dependants':int, 
    'device': str, 
    'urban_city':str,
    'loan_usage':str
}
```

```python
creditDf = credit.astype(convert_dictionary)
```

```python
creditDf['categoria'] = creditDf['loan_amount'].apply(lambda x: 'Bajo'
                                                   if x< 1500
                                                   else ('Medio' if ( x>=1500 and x < 2000)
                                                        else 'Alto'))
```

```python
creditDf['paid_date'] = pd.to_datetime(creditDf['paid_date'])
creditDf['due_date']  = pd.to_datetime(creditDf['due_date'])
creditDf['diasVencido'] = creditDf['paid_date'] - creditDf['due_date']
creditDf = creditDf.dropna(how = 'any')
creditDf['diasVencido'] = pd.to_numeric(creditDf['diasVencido'].dt.days)
```

```python
creditDf.head()
```

```python
sns.stripplot(data=creditDf,
              y='loan_amount',
              x='loan_usage',
              jitter= True)
```

A familiar style of plot that accomplishes this goal is a bar plot. In seaborn, the barplot() function operates on a full dataset and applies a function to obtain the estimate (taking the mean by default). When there are multiple observations in each category, it also uses bootstrapping to compute a confidence interval around the estimate, which is plotted using error bars


Las barras de error son representaciones grficas de la variabilidad de los datos y se utilizan en los grficos para indicar el error o la incertidumbre en una medicin notificada. Dan una idea general de la precisin de una medicin o, por el contrario, de lo lejos que puede estar el valor real del informe. Las barras de error suelen representar una desviacin estndar de la incertidumbre, un error estndar o un intervalo de confianza determinado.
